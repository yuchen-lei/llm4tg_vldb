```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# drawing the marco f1
models = ["GPT-3.5", "GPT-4", "RF", "MLP"]
accuracy = [0.7750, 0.8057259713701431, 0.9118, 0.8947]
macro_precision = [0.7042472735893788, 0.718166853036229, 0.9169, 0.9004]
macro_recall = [0.8631534666719143, 0.8781120794341932, 0.9115, 0.8955]
macro_f1 = [0.6797625130407182, 0.6902858299161718, 0.9135, 0.8975]

bar_width = 0.2  # width of the bars
index = np.arange(len(models))

# Create the bar chart
fig, ax = plt.subplots()

color_palette = ["#3498db", "#e74c3c", "#2ecc71", "#f1c40f"]

bar1 = ax.bar(index, accuracy, bar_width, label="accuracy", color=color_palette[0])
bar2 = ax.bar(
    index + bar_width,
    macro_precision,
    bar_width,
    label="macro-precision",
    color=color_palette[1],
)
bar3 = ax.bar(
    index + 2 * bar_width,
    macro_recall,
    bar_width,
    label="macro-recall",
    color=color_palette[2],
)
bar4 = ax.bar(
    index + 3 * bar_width, macro_f1, bar_width, label="macro-f1", color=color_palette[3]
)

# Add some text for labels, title and custom x-axis tick labels, etc.
# ax.set_xlabel('Models')
# ax.set_ylabel('Scores')
# ax.set_title("Feature based models comparison")
ax.set_xticks(index + bar_width + bar_width / 2)
ax.set_yticks(np.arange(0, 1.1, 0.1))
ax.set_ylim(0, 1.2)
ax.set_xticklabels(models)
# ax.legend(bbox_to_anchor=(1, 1), loc='upper left')
ax.legend()

plt.savefig("models_comparison_feature_based.png", bbox_inches="tight")
plt.show()
```
